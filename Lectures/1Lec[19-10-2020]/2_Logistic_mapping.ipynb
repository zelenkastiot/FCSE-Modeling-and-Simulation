Логистичкото пресликување е модел за раст на број на единки во некоја заедница од животни, или луѓе. Дадено е со еднодимензионален дискретен динамички систем со единствен параметар ламбда
$x_i = Lambda * x_{i-1} * (1 - x_{i-1})$.
Прво се решава моделот еднаш за да се види дали работи добро со разни почетни услови и вредности на параметарот.

```python
import numpy as np
import matplotlib.pyplot as plt

brIteracii = 100 # kolku chekori da napravi

Lambda = 1.1

x = np.zeros((brIteracii)) # br na edinki vi sekoj chekor gi chuva vo vektor

x[0] = 0.2 # pocheten uslov
for i in range(1,brIteracii):
    x[i] = Lambda * x[i - 1] * (1.0 - x[i - 1])
    
plt.plot(range(brIteracii), x,"bs")
plt.xlabel("chekor")
plt.ylabel("naselenie/br. edinki")
plt.show()    
```

Потоа може да се види како зависи поведението од почетниот услов

```python
import numpy as np
import matplotlib.pyplot as plt

brIteracii = 100 # kolku chekori da napravi

Lambda = 1.5

x = np.zeros((brIteracii)) # br na edinki vi sekoj chekor gi chuva vo vektor
y = np.zeros((brIteracii)) 

x[0] = 0.2 # pocheten uslov
y[0] = 0.5 
for i in range(1,brIteracii):
    x[i] = Lambda * x[i - 1] * (1.0 - x[i - 1])
    y[i] = Lambda * y[i - 1] * (1.0 - y[i - 1])    
    
plt.plot(range(brIteracii), x,"bs")
plt.plot(range(brIteracii), y,"ro")

plt.xlabel("chekor")
plt.ylabel("naselenie/br. edinki")
plt.show()    
```

Може да се види и зависноста од параметарот ламбда

```python
import numpy as np
import matplotlib.pyplot as plt

brIteracii = 100 # kolku chekori da napravi

Lambda1 = 1.1
Lambda2 = 1.2

x = np.zeros((brIteracii)) # br na edinki vi sekoj chekor gi chuva vo vektor
y = np.zeros((brIteracii)) 

x[0] = 0.2 # pocheten uslov
y[0] = 0.2 
for i in range(1,brIteracii):
    x[i] = Lambda1 * x[i - 1] * (1.0 - x[i - 1])
    y[i] = Lambda2 * y[i - 1] * (1.0 - y[i - 1])    
    
plt.plot(range(brIteracii), x,"bs")
plt.plot(range(brIteracii), y,"ro")

plt.xlabel("chekor")
plt.ylabel("naselenie/br. edinki")
plt.show()    
```

Личи како секогаш да завршува во иста точка која што зависи од Ламбда (конвергира). Ако побараме решение за x_{i} = x_{i-1} = x, следи  x = lambda * x * (1 - x), односно дека во бескрај x = 1 - 1/lambda.  Проверка дали е тоа во ред

```python
import numpy as np
import matplotlib.pyplot as plt

brIteracii = 100 # kolku chekori da napravi

Lambda = 1.5 # da se vidi za razni lambda

x = np.zeros((brIteracii)) # br na edinki vi sekoj chekor gi chuva vo vektor
razliki = np.zeros((brIteracii)) # br na edinki vi sekoj chekor gi chuva vo vektor

x[0] = 0.2 # pocheten uslov
razliki[0] = x[0] - (1.0 - 1.0 / Lambda)
for i in range(1,brIteracii):
    x[i] = Lambda * x[i - 1] * (1.0 - x[i - 1])
    razliki[i] = x[i] - (1.0 - 1.0 / Lambda)    

plt.plot(range(brIteracii), razliki,"ro")

plt.xlabel("chekor")
plt.ylabel("naselenie/br. edinki")
plt.show()    
```
