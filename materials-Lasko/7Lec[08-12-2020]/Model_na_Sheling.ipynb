Моделот на Шелинг за разграничување на населението од различни раси се состои од агенти (семејства од различни раси) кои што се населени на квадратна решетка. Ако процентот на соседи од иста раса е помал од некој критичен, тогаш семејството се иселува на случајно одбрано место

```python

import numpy as np
import random
import matplotlib.pyplot as plt

def Zadovolni(Grad, prag):
    zadovolniBroj = 0
    GoleminaNaGrad = len(Grad)
    for red in range(GoleminaNaGrad):
        for kolona in range(GoleminaNaGrad):
            pripadnost = Grad[red, kolona]
            if pripadnost > 0:
                sonarodnici = 0
                sosedi = 0
                # broenje na sosedite i sonarodnicite vo sosedstvoto
                if red > 0:
                    if Grad[red - 1, kolona] == pripadnost:
                        sonarodnici = sonarodnici + 1
                    if Grad[red - 1, kolona] > 0:
                        sosedi = sosedi + 1
                        
                if red > 0 and kolona > 0:
                    if Grad[red - 1, kolona - 1] == pripadnost: 
                        sonarodnici = sonarodnici + 1    
                    if Grad[red - 1, kolona - 1] > 0: 
                        sosedi = sosedi + 1
                           
                if red > 0 and kolona < GoleminaNaGrad - 1:        
                    if Grad[red - 1, kolona + 1] == pripadnost: 
                        sonarodnici = sonarodnici + 1                                   
                    if Grad[red - 1, kolona + 1] > 0: 
                        sosedi = sosedi + 1
                        
                if red < GoleminaNaGrad - 1:
                    if Grad[red + 1, kolona] == pripadnost:            
                        sonarodnici = sonarodnici + 1    
                    if Grad[red + 1, kolona] > 0:            
                        sosedi = sosedi + 1
        
                if kolona > 0 and red < GoleminaNaGrad - 1:
                    if Grad[red + 1, kolona - 1] == pripadnost: 
                        sonarodnici = sonarodnici + 1    
                    if Grad[red + 1, kolona - 1] > 0: 
                        sosedi = sosedi + 1
        
                if red < GoleminaNaGrad - 1 and kolona < GoleminaNaGrad - 1:
                    if Grad[red + 1, kolona + 1] == pripadnost:
                        sonarodnici = sonarodnici + 1    
                    if Grad[red + 1, kolona + 1] > 0:
                        sosedi = sosedi + 1
        
                if kolona > 0:
                    if Grad[red, kolona - 1] == pripadnost: 
                        sonarodnici = sonarodnici + 1    
                    if Grad[red, kolona - 1] > 0: 
                        sosedi = sosedi + 1
        
                if kolona < GoleminaNaGrad - 1:
                    if Grad[red, kolona + 1] == pripadnost:
                        sonarodnici = sonarodnici + 1    
                    if Grad[red, kolona + 1] > 0:
                        sosedi = sosedi + 1

                if sosedi == 0:
                    zadovolniBroj = zadovolniBroj + 1
                else:
                    if 1.0 * sonarodnici / sosedi > prag:
                        zadovolniBroj = zadovolniBroj + 1            
    return 1.0 * zadovolniBroj / (GoleminaNaGrad * GoleminaNaGrad)                    



GoleminaNaGrad = 30
PragNaPodnoslivost = 0.45
BrojNaChekori = 30000

Grad = np.zeros((GoleminaNaGrad, GoleminaNaGrad))

ProcentCrveni = 0.45 # oznacheni so 1
ProcentBeli = 0.25 # oznacheni so 2
ProcentSini = 0.2  # oznacheni so 3

for i in range(GoleminaNaGrad):
    for j in range(GoleminaNaGrad):
        SluchaenIzbor = random.random()
        if SluchaenIzbor < ProcentCrveni:
            Grad[i, j] = 1
        elif SluchaenIzbor < ProcentCrveni + ProcentBeli:
            Grad[i, j] = 2
        elif SluchaenIzbor < ProcentCrveni + ProcentBeli + ProcentSini:
            Grad[i, j] = 3
        
            
plt.imshow(Grad, cmap='hot', interpolation='nearest')            
plt.colorbar()
plt.show()
print(Zadovolni(Grad, PragNaPodnoslivost))


for i in range(BrojNaChekori):
    red = random.randint(0, GoleminaNaGrad - 1)
    kolona = random.randint(0, GoleminaNaGrad - 1)
    pripadnost = Grad[red, kolona]
    sosedi = 0
    if pripadnost > 0:
        sonarodnici = 0
        sosedi = 0
        # broenje na sosedite i sonarodnicite vo sosedstvoto
        if red > 0:
            if Grad[red - 1, kolona] == pripadnost:
                sonarodnici = sonarodnici + 1
            if Grad[red - 1, kolona] > 0:
                sosedi = sosedi + 1
                
        if red > 0 and kolona > 0:
            if Grad[red - 1, kolona - 1] == pripadnost: 
                sonarodnici = sonarodnici + 1    
            if Grad[red - 1, kolona - 1] > 0: 
                sosedi = sosedi + 1
                   
        if red > 0 and kolona < GoleminaNaGrad - 1:        
            if Grad[red - 1, kolona + 1] == pripadnost: 
                sonarodnici = sonarodnici + 1                                   
            if Grad[red - 1, kolona + 1] > 0: 
                sosedi = sosedi + 1
                
        if red < GoleminaNaGrad - 1:
            if Grad[red + 1, kolona] == pripadnost:            
                sonarodnici = sonarodnici + 1    
            if Grad[red + 1, kolona] > 0:            
                sosedi = sosedi + 1

        if kolona > 0 and red < GoleminaNaGrad - 1:
            if Grad[red + 1, kolona - 1] == pripadnost: 
                sonarodnici = sonarodnici + 1    
            if Grad[red + 1, kolona - 1] > 0: 
                sosedi = sosedi + 1

        if red < GoleminaNaGrad - 1 and kolona < GoleminaNaGrad - 1:
            if Grad[red + 1, kolona + 1] == pripadnost:
                sonarodnici = sonarodnici + 1    
            if Grad[red + 1, kolona + 1] > 0:
                sosedi = sosedi + 1

        if kolona > 0:
            if Grad[red, kolona - 1] == pripadnost: 
                sonarodnici = sonarodnici + 1    
            if Grad[red, kolona - 1] > 0: 
                sosedi = sosedi + 1

        if kolona < GoleminaNaGrad - 1:
            if Grad[red, kolona + 1] == pripadnost:
                sonarodnici = sonarodnici + 1    
            if Grad[red, kolona + 1] > 0:
                sosedi = sosedi + 1

    if sosedi > 0: # za da ne delime so nula   
        if 1.0 * sonarodnici / sosedi < PragNaPodnoslivost: # premesti go
            redNov = random.randint(0, GoleminaNaGrad - 1)
            kolonaNova = random.randint(0, GoleminaNaGrad - 1)
            while Grad[redNov, kolonaNova] > 0:
                redNov = random.randint(0, GoleminaNaGrad - 1)
                kolonaNova = random.randint(0, GoleminaNaGrad - 1)
            Grad[redNov, kolonaNova] = pripadnost
            Grad[red, kolona] = 0
            
plt.imshow(Grad, cmap='hot', interpolation='nearest')            
plt.colorbar()
plt.show()        
    
print(Zadovolni(Grad, PragNaPodnoslivost))

```
