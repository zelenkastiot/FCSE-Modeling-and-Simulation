Логистичка равенка со шум дадена со $dN=rN\left(1-\frac{N}{K}\right)dt + \sigma N dW$

```python
import numpy as np
import matplotlib.pyplot as plt

r = 0.1 # stapka na prirast
sigma = 0.01 # varijansa na raspredelbata na shumot
K = 1000 # kapacitet na zhivealishteto
chekori = 20000
deltaT = 0.001
edinki = np.zeros(chekori)
vreminja = np.zeros(chekori)

edinki[0] = 5 # pochetna naselenost
for i in range(chekori - 1):
    vreminja[i + 1] = 1.0 * (i + 1) * deltaT
    edinki[i + 1] = edinki[i] * (1.0 + r * deltaT * (1.0 - edinki[i] / K) + sigma * np.sqrt(deltaT)* np.random.normal())

plt.plot(vreminja, edinki)
plt.show()   
```

Модел на цена на акција според Википедија дадена со Геометриско Брауново движење кое што е дадено со $X(t) = X_0 e^{(\mu - \sigma^2/2)t + \sigma W_t}$, каде што $W_t$ е Винеров процес

```python

#python code for the plot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

mu=1
n=50
dt=0.1
x0=100
x=pd.DataFrame()
np.random.seed(1)

for sigma in np.arange(0.8,2,0.2):
    step=np.exp((mu-sigma**2/2)*dt)*np.exp(sigma*np.random.normal(0,dt,(1,n)))
    temp=pd.DataFrame(x0*step.cumprod())
    x=pd.concat([x,temp],axis=1)

x.columns=np.arange(0.8,2,0.2)
plt.plot(x)
plt.legend(x.columns)
plt.xlabel('t')
plt.ylabel('X')
plt.title('Геометриско Брауново движење\n за ми=1')
plt.show()
```

Симулација на модел на цена на акција

```python
import numpy as np
import matplotlib.pyplot as plt

r = 0.01 # kamata
sigma = 0.03 # varijansa na raspredelbata na shumot

chekori = 20000
deltaT = 0.0001
ceni = np.zeros(chekori)
vreminja = np.zeros(chekori)

ceni[0] = 1.0 # pochetna cena
for i in range(chekori - 1):
    vreminja[i + 1] = 1.0 * (i + 1) * deltaT
    ceni[i + 1] = ceni[i] * (1.0 + r * deltaT + sigma * np.sqrt(deltaT) * np.random.normal())

plt.plot(vreminja, ceni)
plt.xlabel('t')
plt.ylabel('X')

plt.show()    
#print(ceni)
```
